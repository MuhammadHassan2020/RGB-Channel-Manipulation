```python
# sample dataset
x_train = np.array([
    [1,2],
    [8,8],
    [3,3],
    [6,5],
    [7,8],
    [2,3]
])
y_train = np.array([0,1,0,1,1,0])  # labels for the training data
```

```python
def minkowski_distance(point1, point2, p=2):
  return np.sum(np.abs(point1-point2)**p)**(1/p)
```

```python
# KNN

def knn(X_train, y_train, X_test, k=3, p=2):
  predictions = []
  for test_point in X_test:
    distances = [minkowski_distance(test_point, x_train, p) for x_train in X_train]
    print(f'Distance: {distances}')
    k_indices = np.argsort(distances)[:k]
    print(f'K indices: {k_indices}')
    k_nearest_labels = [y_train[i] for i in k_indices]
    print(f'K nearest labels: {k_nearest_labels}')
    most_common = Counter(k_nearest_labels).most_common(1)
    print(f'Most common: {most_common}')
    predictions.append(most_common[0][0])
  return predictions
```

```python
k = 3
p = 2
predictions = knn(x_train, y_train, x_test, k, p)
print("Predictions: ",predictions)
```
